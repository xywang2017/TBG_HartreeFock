```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
# np.NaN means non-convergent or no clear magnetic subbands
gaps2 = np.array([[i/2 for i in range(7)],
         [20.639,13.65,14.34,13.00,13.04,np.NaN,np.NaN]])

gaps3 = np.array([[i/3 for i in range(10)],
         [18.83,10.94,12.86,np.NaN,0.82,16.37,15.87,21.90,14.81,19.91]])

gaps4 = np.array([[i/4 for i in range(13)],
         [16.93,8.24,10.75,10.52,np.NaN,11.23,15.0,17.56,1.44,20.00,16.43,1.76,20.51]])

gaps5 = np.array([[i/5 for i in range(16)],
         [16.97,7.56,10.29,10.03,np.NaN,8.53,12.08,14.64,18.80,np.NaN,np.NaN,19.27,17.52,np.NaN,np.NaN,20.27]])

gaps6 = np.array([[i/6 for i in range(18)],
         [16.7,6.94,np.NaN,9.48,np.NaN,np.NaN,7.98,13.56,14.02,1.15,np.NaN,1.47,9.49,18.29,18.10,1.39,np.NaN,np.NaN]])

gaps8 = np.array([[i/8 for i in range(21)],
         [15.41,2.57,5.15,0.79,0.97,np.NaN,np.NaN,8.06,3.81,0.28,9.1,1.90,0.73,np.NaN,0.66,0.47,3.51,12.58,3.09,1.49,0.78]])

gaps10 = np.array([[i/10 for i in range(16)],
        [15.61,4.41,6.78,6.63,np.NaN,np.NaN,np.NaN,np.NaN,np.NaN,np.NaN,5.82,13.05,10.49,2.62,1.88,np.NaN]])
```

```python
fig = plt.figure(figsize=(6,4))
rat = 3
plt.axvline(0,ls=":",c="gray")
plt.axvline(1,ls=":",c="gray")
plt.axvline(2,ls=":",c="gray")
plt.axvline(3,ls=":",c="gray")
plt.scatter(gaps2[0], np.ones(gaps2.shape[1])/2, s=gaps2[1]**2/rat)
plt.scatter(gaps3[0], np.ones(gaps3.shape[1])/3, s=gaps3[1]**2/rat)
plt.scatter(gaps4[0], np.ones(gaps4.shape[1])/4, s=gaps4[1]**2/rat)
plt.scatter(gaps5[0], np.ones(gaps5.shape[1])/5, s=gaps5[1]**2/rat)
plt.scatter(gaps6[0], np.ones(gaps6.shape[1])/6, s=gaps6[1]**2/rat)
plt.scatter(gaps8[0], np.ones(gaps8.shape[1])/8, s=gaps8[1]**2/rat)
plt.scatter(gaps10[0], np.ones(gaps10.shape[1])/10, s=gaps10[1]**2/rat)
plt.xlabel(r"$\nu$")
plt.ylabel(r"$\phi/\phi_0$")

plt.xlim(-0.3,4)
plt.ylim(0,0.6)
plt.show()
```
